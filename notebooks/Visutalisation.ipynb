```python
import os
import random
from matplotlib import pyplot as plt
from PIL import Image

base_dir = '../dogs_dataset/images'

# Liste des races disponibles
races = os.listdir(base_dir)
print(f"Nombre de races trouv√©es : {len(races)}")

# Choisir quelques races al√©atoirement
sample_races = random.sample(races, 3)

plt.figure(figsize=(12, 6))

for i, race in enumerate(sample_races, 1):
    race_dir = os.path.join(base_dir, race)
    images = os.listdir(race_dir)
    # Choisir une image al√©atoire dans la race
    img_name = random.choice(images)
    img_path = os.path.join(race_dir, img_name)
    
    # Ouvrir l'image
    img = Image.open(img_path)
    
    # Afficher
    plt.subplot(1, 3, i)
    plt.imshow(img)
    plt.title(race)
    plt.axis('off')

plt.show()
```
